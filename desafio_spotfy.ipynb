```python
import pandas as pd

df = pd.read_csv('/content/top50.csv', encoding='latin-1', index_col=0)

df.head()
```

```python
print(f'For each one of the {df.shape[0]} songs in the dataset we have {df.shape[1]} atributes that are :')
for column in df.columns:
  print("  >>", column)
```

```python
# I will improve the column name so it will be easier to call them when we need:
new_names = {
    'Track.Name':'music',
    'Artist.Name': 'artist',
    'Genre': 'genre',
    'Beats.Per.Minute':'beats_per_minute',
    'Energy': 'energy',
    'Danceability': 'danceability',
    'Loudness..dB..':'loudness_db',
    'Liveness': 'liveness', 
    'Valence.': 'valence', 
    'Length.': 'length',
    'Acousticness..': 'acousticness', 
    'Speechiness.': 'speechiness', 
    'Popularity': 'popularity'
    }
df.rename(columns=new_names, inplace=True)
df.tail()
```
